```python
import json
def whatsapp(url):
    File=open(url)
    lst=[]
    conversation=dict()
    iD=0
    all_data=dict()

    for line in File:
        flag=None
        try:
            if ('<×”××“×™×”' in line or "×”×•×¡×™×£/×”" in line or " ××•×¦×¤× ×•×ª " in line or "× ×•×¦×¨×” ×¢×œ ×™×“×™" in line):
                continue #skip unnecessary raws
            sms=dict() 
            datetime=line.index('-')
            date=line[0:datetime]   
            text=line[line.index(':',datetime) +1:]
            phone=line[datetime+1:line.index(':',datetime)]
            for key,value in conversation.items():
                if value==phone:
                    flag=key
            if flag is not None:
                phone=flag
            else:
                conversation[iD]=phone
                phone=iD
                iD+=1
            sms['datetime'] = date
            sms['id'] = phone
            sms['text'] = text
            lst.append(sms)
        except:
            lst[-1]['text'] += line #takes the lines that needs to be with privious raw



    #Metadata
    File=open("/Users/lidoriluz/Downloads/ï¿½×¦××˜ WhatsApp ×¢× ×™×•× ×”×•×œ×“×ª ×‘× ×•×ª ×œ× ×•×™×”.txt",encoding='utf8')
    for line in File:
        if  "× ×•×¦×¨×” ×¢×œ ×™×“×™" not  in line:
            continue
        metadata = dict()
        group = line.split('"')
        chat_name=group[1]
        metadata["chat_name"] = group[1]
        creat_date = line.split('-')
        metadata["creation_date"]=creat_date[0]
        metadata["num_of_participants"]=iD+1
        creator_index=line.index('×™×“×™')
        metadata["creator"]=line[creator_index+6:creator_index+22]



    all_data["messages"] = lst 
    all_data["metadata"] = metadata
    
    #create JSON:
    file_name=chat_name + '.txt'
    file= open(file_name, 'w', encoding='utf-8') 
    json.dump(all_data, file, ensure_ascii=False, indent=4)
    file.close()
    
    return(all_data)

#activates fun.
whatsapp("/Users/lidoriluz/Downloads/ï¿½×¦××˜ WhatsApp ×¢× ×™×•× ×”×•×œ×“×ª ×‘× ×•×ª ×œ× ×•×™×”.txt")
```
